```python
########## 2回目
import pandas as pd
import numpy as np

train = pd.read_csv('./data/train.csv', index_col=0)
test = pd.read_csv('./data/test.csv', index_col=0)
sample_submit = pd.read_csv('./data/sample_submission.csv', index_col=0)
print(sample_submit)
from sklearn.model_selection import train_test_split
# from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import f1_score

def label_encoding(data):
    '''
    カテゴリーをLabelエンコーディング
    '''
    LE = LabelEncoder()
    LE.fit_transform(data.values)
    return LE.fit_transform(data.values)

# nanの行列を削除
df2 = train.dropna(how='any').dropna(how='any', axis=1)
# print(df2)
# print(df2.isnull().sum())

# df2 を変換
df2["HomePlanet"] = label_encoding(df2["HomePlanet"])
df2["CryoSleep"] = label_encoding(df2["CryoSleep"])
df2["Cabin"] = label_encoding(df2["Cabin"])
df2["Destination"] = label_encoding(df2["Destination"])
df2["VIP"] = label_encoding(df2["VIP"])
df2["Name"] = label_encoding(df2["Name"])

# test を変換
test["HomePlanet"] = label_encoding(test["HomePlanet"])
test["CryoSleep"] = label_encoding(test["CryoSleep"])
test["Cabin"] = label_encoding(test["Cabin"])
test["Destination"] = label_encoding(test["Destination"])
test["VIP"] = label_encoding(test["VIP"])
test["Name"] = label_encoding(test["Name"])

# 訓練データとテストデータに分割
# stratifyを設定して目的変数の分布を保ったまま分割
train, valid = train_test_split(df2, test_size=0.2, stratify=df2['Transported'], random_state=82)

# 目的変数とそれ以外に学習用データを分割
x_train = train.drop(columns=['Transported']) 
y_train = train['Transported']
x_valid = valid.drop(columns=['Transported']) 
y_valid = valid['Transported']
x_test = test

# モデルを作成して訓練
# model = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000)
model = HistGradientBoostingClassifier(max_iter=1000)
model.fit(x_train, y_train)

# validの特徴量で予測
valid_predictions = model.predict(x_valid)
valid_f1 = f1_score(y_valid, valid_predictions, average='macro')
print(f"Validation F1 Score (Macro): {valid_f1}")

# 予測結果をCSVファイルとして保存
pred = model.predict(x_test)
print(pred)
# sample_submit[0] = ['PassengerId', 'Transported']
sample_submit[1] = pred
sample_submit.to_csv('./submit/submit2.csv')
```
