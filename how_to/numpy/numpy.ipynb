# 配列の計算 

```python
######## 四則演算
import numpy as np
arr1 = np.array([100,200,300])
arr2 = np.array([30,40,50])

# 配列の加算
print(np.add(arr1,arr2))

# 配列の減算
print(np.subtract(arr1,arr2))

# 配列の乗算
print(np.multiply(arr1,arr2))

# 配列の除算
print(np.divide(arr1,arr2))
```

```python
######## 指数関数と対数関数
import numpy as np
arr = np.array([0.1,1,np.e])

# 指数関数
print(np.exp([0.1, 1, np.e]))

# 対数関数
print(np.log([0.1, 1, np.e]))
```

```python
######## 三角関数(sin, cos, tan)
import numpy as np
import matplotlib.pyplot as plt
# -π〜πの等差数列の作成
arr = np.linspace(-np.pi, np.pi, 100)

def wave_plot(arr, result, ylabel):
    plt.plot(arr, result)
    plt.xlabel('Angle [rad]')
    plt.ylabel(ylabel)
    plt.axis('tight')
    plt.show()

# 正弦の波形のプロット
wave_plot(arr, np.sin(arr), ylabel='sin(x)')
# 余弦の波形のプロット
wave_plot(arr, np.cos(arr), ylabel='cos(x)')
# 正接の波形のプロット
wave_plot(arr, np.tan(arr), ylabel='tan(x)')
```

```python
######## 平方根
import numpy as np
arr = np.array([0,100,10000])

# 平方根の算出
print(np.sqrt(arr))
```

```python
######## 内積
import numpy as np
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])

# 内積の算出
print(np.dot(arr1,arr2))
```

```python
######## データを複製する
import numpy as np

arr = np.eye(3,3)
new_arr = arr # 複製後の配列を変更すると複製元の配列も変更される
view_arr = arr.view() # 複製後の配列を変更すると複製元の配列も変更される
copy_arr = arr.copy() # 複製元と複製後の配列でお互いに影響を与えることがない

print(arr)
print(new_arr)
print(view_arr)
print(copy_arr)

print("-----")

view_arr[0][0] = 2
print(arr)
print(new_arr)
print(view_arr)
print(copy_arr)

print("-----")

copy_arr[0][0] = 3
print(arr)
print(new_arr)
print(view_arr)
print(copy_arr)
```

```python
######## 配列の変形(1) 転置
import numpy as np
np.random.seed(seed=0)

# 配列の生成
arr = np.random.randint(0,100,(2,3))
print(arr)

# 配列の転置
arr1 = np.transpose(arr)
print(arr1)

arr2 = arr.T
print(arr2)
```

```python
######## 配列の変形(2) 平坦化
import numpy as np
np.random.seed(seed=0)

# 配列の生成
arr = np.random.randint(0,100,(2,3))
print(arr)

# 配列の平坦化
arr1 = arr.ravel()
print(arr1)

arr2 = arr.flatten()
print(arr2)
```

```python
######## 配列の変形(3) 万能な変形
import numpy as np
np.random.seed(seed=0)
# reshape関数で指定する配列の要素数は元の配列の要素数から変更することができない
# resize関数はreshape関数と異なり、形状を限定せずに使用することができる


# 配列の生成
arr = np.random.randint(0,100,(2,3))
print(arr)

# 2行3列の配列を3行2列の配列に変形する
print('2行3列　->　3行2列')
arr1 = np.reshape(arr,(3,2))
print(arr1)

arr2 = np.resize(arr,(3,2))
print(arr2)

# 2行3列の配列を2行2列の配列に変形する
print('2行3列　->　2行2列')
arr3 = np.resize(arr,(2,2))
print(arr3)

# 2行3列の配列を4行5列の配列に変形する
print('2行3列　->　4行5列')
arr4 = np.resize(arr,(4,5)) 
print(arr4)
```

# 統計

```python
######## 最大値の算出
import numpy as np

arr = np.array([
 [42120, 42330, 41700, 41830,610000], 
 [43000, 43220, 42540, 42660,448400],
 [43940, 43970, 43270, 43270,339900], 
 [13990, 14030, 13790, 13850,765500], 
 [13720, 13840, 13600, 13800,511500], 
 [14050, 14050, 13700, 13720,559100]])

#arrの1行目をスライスする
arr_initial = arr[0]
print(arr_initial)

# 行の最大値を求める
arr_initial_max = arr_initial.max() 
print(arr_initial_max)

#各行の最大値を全て求める
arr_rmax = arr.max(axis=1) 
print(arr_rmax)
```

```python
######## 最小値の算出
import numpy as np

arr = np.array([
 [42120, 42330, 41700, 41830,610000], 
 [43000, 43220, 42540, 42660,448400],
 [43940, 43970, 43270, 43270,339900], 
 [13990, 14030, 13790, 13850,765500], 
 [13720, 13840, 13600, 13800,511500], 
 [14050, 14050, 13700, 13720,559100]])

#arrの1行目をスライスする
arr_initial = arr[0]
print(arr_initial)

# 1行目の最小値を求める
arr_initial_min = arr_initial.min() 
print(arr_initial_min)

# 各行の最小値を全て求める
arr_rmin = arr.min(axis=1) 
print(arr_rmin)
```

```python
######## 平均、中央値の算出
import numpy as np

arr = np.array([
 [42120, 42330, 41700, 41830,610000], 
 [43000, 43220, 42540, 42660,448400],
 [43940, 43970, 43270, 43270,339900], 
 [13990, 14030, 13790, 13850,765500], 
 [13720, 13840, 13600, 13800,511500], 
 [14050, 14050, 13700, 13720,559100]])

#各行の平均値
arr_mean = np.mean(arr,axis=1)

#3行目のデータをスライス
arr3 = arr[2]

arr3_mean = np.mean(arr3)
arr3_median = np.median(arr3)

print(arr_mean)
print('---arr3の平均値と中央値---')
print(arr3_mean)
print(arr3_median)
```

```python
######## 分散、標準偏差の算出
import numpy as np

arr = np.array([
 [42120, 42330, 41700, 41830,610000], 
 [43000, 43220, 42540, 42660,448400],
 [43940, 43970, 43270, 43270,339900], 
 [13990, 14030, 13790, 13850,765500], 
 [13720, 13840, 13600, 13800,511500], 
 [14050, 14050, 13700, 13720,559100]])

print('---配列の分散---')
print(np.var(arr,axis=0))
print(np.var(arr,axis=1))
print('---配列の標準偏差---')
print(np.std(arr,axis=0))
print(np.std(arr,axis=1))
```

```python
######## 相関の算出
import numpy as np

arr = np.array([
 [42120, 42330, 41700, 41830,610000], 
 [43000, 43220, 42540, 42660,448400],
 [43940, 43970, 43270, 43270,339900], 
 [13990, 14030, 13790, 13850,765500], 
 [13720, 13840, 13600, 13800,511500], 
 [14050, 14050, 13700, 13720,559100]])

#相関係数の計算
coef1 = np.corrcoef(arr.T[0],arr.T[1])
print(coef1)
```

```python
######## 累積和の算出
import numpy as np

arr = np.array([
 [42120, 42330, 41700, 41830,610000], 
 [43000, 43220, 42540, 42660,448400],
 [43940, 43970, 43270, 43270,339900], 
 [13990, 14030, 13790, 13850,765500], 
 [13720, 13840, 13600, 13800,511500], 
 [14050, 14050, 13700, 13720,559100]])
arr3 = arr[2]

#配列の合計値
arr_sum = np.sum(arr3)

#配列の累積和
arr_cumsum = np.cumsum(arr3)

print(arr3)
print(arr_sum)
print(arr_cumsum)
```
