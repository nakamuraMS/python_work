# グラフの作成

```python
'''
グラフの種類と用途
折れ線グラフ 	数量の時系列での変化の確認 	売上の月別推移
棒グラフ 	データ間の数量の大きさの比較 	部門別の売上の比較
散布図 	2種類のデータの相関の確認 	商品の売上と気温の関係
ヒストグラム 	データのばらつき（分布）の確認 	製品の品質の確認
箱ひげ図 	データのばらつき（分布）の確認 	製品の品質の確認
ヒートマップ 	データの偏りの確認 	年度×月別の売上の傾向
'''

# matplotlib.pyplotのインポート
import matplotlib.pyplot as plt

# seabornのインポート
import seaborn as sns

# アヤメデータの読み込み
iris = sns.load_dataset("iris")

# スタイルの設定
sns.set_style("white")
iris.plot(kind='scatter', x="sepal_length", y="sepal_width")
plt.show()

sns.set_style("darkgrid")
iris.plot(kind='scatter', x="sepal_length", y="sepal_width")
plt.show()
```

```python
######## 折れ線グラフ
# matplotlib.pyplotのインポート
import matplotlib.pyplot as plt

# seabornのインポート
import seaborn as sns

# データの読み込み
fmri = sns.load_dataset("fmri")

sns.set_style("whitegrid")

fig, ax = plt.subplots(1,4, figsize=(16,3), sharey=True)
sns.lineplot(x="timepoint", y="signal", data=fmri, ax=ax[0])
sns.lineplot(x="timepoint", y="signal", data=fmri, ax=ax[1], ci=50)
sns.lineplot(x="timepoint", y="signal", data=fmri, ax=ax[2], hue="event")
sns.lineplot(x="timepoint", y="signal", data=fmri, ax=ax[3], hue="region", style="event")

plt.show()
```

```python
######## 棒グラフ
# matplotlib.pyplotのインポート
import matplotlib.pyplot as plt

# seabornのインポート
import seaborn as sns

# データの読み込み
titanic = sns.load_dataset("titanic")
plt.rcParams['font.family'] = 'MigMix 1M' # これで日本語文字化けをなくしている

# figureオブジェクトとaxesオブジェクトの作成
fig, ax = plt.subplots(1,4, figsize=(16,3), sharey=True)

# グラフの作成
sns.barplot(x="class", y="survived", data=titanic, ax=ax[0])
sns.barplot(x="class", y="survived", data=titanic, ax=ax[1], order=["Third", "Second", "First"])
sns.barplot(x="class", y="survived", data=titanic, ax=ax[2], ci=None, palette="coolwarm")
sns.barplot(x="class", y="survived", data=titanic, ax=ax[3], hue="sex")

ax[0].set(title="デフォルト")
ax[1].set(title="表示順を変更")
ax[2].set(title="エラーバンド非表示 & パレット変更")
ax[3].set(title="性別で分割")

plt.show()
```

```python
######## 箱ひげ図
# matplotlib.pyplotのインポート
import matplotlib.pyplot as plt

# seabornのインポート
import seaborn as sns

# データの読み込み
iris = sns.load_dataset("iris")

# figureオブジェクトとaxesオブジェクトの作成
fig, ax = plt.subplots(2, 3, figsize=(15,10), sharey=True)

# グラフの作成
sns.boxplot(x="species", y="sepal_length", data=iris, ax=ax[0,0])
sns.stripplot(x="species", y="sepal_length", data=iris, ax=ax[0,1])
sns.violinplot(x="species", y="sepal_length", data=iris, ax=ax[0,2])
sns.boxenplot(x="species", y="sepal_length", data=iris, ax=ax[1,0])
sns.swarmplot(x="species", y="sepal_length", data=iris, ax=ax[1,1])
sns.violinplot(x="species", y="sepal_length", data=iris, ax=ax[1,2], color="whitesmoke", linewidth=0.3)
sns.swarmplot(x="species", y="sepal_length", data=iris, ax=ax[1,2], size=3)

ax[0,0].set_title("boxplot")
ax[0,1].set_title("stripplot")
ax[0,2].set_title("violinplot")
ax[1,0].set_title("boxenplot")
ax[1,1].set_title("swarmplot")
ax[1,2].set_title("violinplot & swarmplot")

plt.show()
```

```python
######## ヒストグラム
# matplotlib.pyplotのインポート
import matplotlib.pyplot as plt

# seabornのインポート
import seaborn as sns

# データの読み込み
iris = sns.load_dataset("iris")

# 描画対象データの抽出
fig, ax = plt.subplots(1,4, figsize=(20,4))

sepal_width = iris["sepal_width"]
sns.distplot(sepal_width, ax=ax[0])
sns.distplot(sepal_width, ax=ax[1], bins=20)
sns.distplot(sepal_width, ax=ax[2], kde=False)
sns.distplot(sepal_width, ax=ax[3], hist=False, rug=True)

ax[0].set_title("hist & kde")
ax[1].set_title("hist (bins=20) & kde")
ax[2].set_title("hist")
ax[3].set_title("kde & rug")

plt.show()
```

```python
######## 散布図
# matplotlib.pyplotのインポート
import matplotlib.pyplot as plt

# seabornのインポート
import seaborn as sns

# データの読み込み
iris = sns.load_dataset("iris")

# figureオブジェクトとaxesオブジェクトの作成
fig, ax = plt.subplots(1,4, figsize=(24,4), sharey=True)

# グラフの作成
sns.scatterplot(x="sepal_length", y="sepal_width", data=iris, ax=ax[0])
sns.scatterplot(x="sepal_length", y="sepal_width", data=iris, ax=ax[1], hue="species")
sns.scatterplot(x="sepal_length", y="sepal_width", data=iris, ax=ax[2], hue="species", style="species")
sns.scatterplot(x="sepal_length", y="sepal_width", data=iris, ax=ax[3], hue="species", style="species", size="petal_length")

plt.show()
```

```python
######## 散布図行列
# matplotlib.pyplotのインポート
import matplotlib.pyplot as plt

# seabornのインポート
import seaborn as sns

# データの読み込み
iris = sns.load_dataset("iris")

# グラフの作成
sns.pairplot(data=iris, hue="species")

# グラフの表示
plt.show()
```

```python
######## ヒートマップ
# matplotlib.pyplotのインポート
import matplotlib.pyplot as plt

# seabornのインポート
import seaborn as sns

# pandasのインポート
import pandas as pd

# データの読み込み
flights = sns.load_dataset("flights")

# figureオブジェクトとaxesオブジェクトの作成
fig, ax = plt.subplots(1, 2, figsize=(18, 6))

# 年と月ごとに乗客を集約した2次元配列データの作成
flights_matrix = pd.pivot_table(data=flights, index="year", columns="month", values="passengers")

sns.heatmap(data=flights_matrix, ax=ax[0])
sns.heatmap(data=flights_matrix, ax=ax[1], cmap='coolwarm', annot=True, fmt="d", vmax=450)
fig.suptitle("ヒートマップ", fontsize=20)

plt.show()
```

```python
######## 複数のカテゴリの比較
# matplotlib.pyplotのインポート
import matplotlib.pyplot as plt

# seabornのインポート
import seaborn as sns

# データの読み込み
tips = sns.load_dataset("tips")

# グラフの作成
sns.catplot(x="day", y="tip", row="sex", col="smoker", hue="time", data=tips, kind="bar", aspect=1.5) 

# グラフの表示
plt.show()
```

```python
######## 散布図と回帰曲線
# matplotlib.pyplotのインポート
import matplotlib.pyplot as plt

# seabornのインポート
import seaborn as sns

# データの読み込み
tips = sns.load_dataset("tips")

# グラフの作成
sns.lmplot(x="total_bill", y="tip", data=tips, hue="smoker")

# グラフの表示
plt.show()
```
